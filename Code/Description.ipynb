# $$\text{QUADRUPED ROBOT}$$

## Some Notation and Physical Parameters

Here, we provide some notation for the robot model:
- $m(kg)$ is the mass of the whole robot (mass of the base) assuming that the legs have zero mass.
- $L(m)$ is the length of the robot body.
- $W(m)$ is the width of the robot body.
- $H(m)$ is the height of the robot body.
- $$

$$
M\ddot{q_b}+C(\dot{q_b})\dot{q_b}+G = J(q_j)^T_{c,b}\lambda
$$

$$
\tau = -J(q_j)^T_{c,j}\lambda
$$

Let define the body state vector $s_b\in \mathbb{R}^{12\times 1}$:

$$
s_b =\begin{bmatrix}q\\ \dot{q}\end{bmatrix}=\begin{bmatrix}
x\\ y\\ z \\ \phi \\ \theta \\ \psi \\ \dot{x} \\ \dot{y} \\ \dot{z} \\ \dot{\phi} \\ \dot{\theta} \\ \dot{\psi}
\end{bmatrix} =\begin{bmatrix}
x\\ y \\ z \\ \phi \\ \theta \\ \psi \\ v_x \\ v_y \\ v_z \\ \omega_\phi \\ \omega_\theta \\ \omega_\psi
\end{bmatrix}
$$

And the joint state vector $s_j=q_j\in \mathbb{R}^{12\times 1}$:
$$
s_j=q_j =\begin{bmatrix}
\phi_{fl_1} \\ \phi_{fl_2} \\ \phi_{fl_3} \\ \phi_{fr_1} \\ \phi_{fr_2} \\ \phi_{fr_3} \\ \phi_{bl_1} \\ \phi_{bl_2} \\ \phi_{bl_3} \\ \phi_{br_1} \\ \phi_{br2_1} \\ \phi_{br_3} \\ 
\end{bmatrix}
$$

Extend the state to become $s=[s_b,s_j]^T\in\mathbb{R^{24\times 1}}$

with $h=\Delta t$, the system eq becomes:

$$
s_b=\begin{bmatrix}q\\ \dot{q}\end{bmatrix}\quad;\quad\dot{s_b}=\begin{bmatrix}\dot{q}\\ \ddot{q}\end{bmatrix}\quad;\quad s=\begin{bmatrix}\dot{q}\\ \ddot{q}\\s_j\end{bmatrix}
$$

$$
M\dot{s_b} = J(s_j)^T_{c,b}\lambda-G-C(s_b)s_b
$$

$$
\dot{s_b} = M^{-1}\left(J(s_j)^T_{c,b}\lambda-G-C(s_b)s_b\right)
$$

take integral both sides:
$$
\int_{t_k}^{t_{k+1}}\dot{s_b} dt = \int_{t_k}^{t_{k+1}}\left\{M^{-1}\left(J(s_j)^T_{c,b}\lambda-G-C(s_b)s_b\right)\right\}dt=\int_{t_k}^{t_{k+1}}f(s_b,s_j)dt
$$

$$
s_{b_{k+1}}-s_{b_{k}}=\frac{h}{2}(f_{k+1}+f_k)
$$

## Optimization variables:
- $s_{b_k}$ for $k=0,1,...,T$
- $s_{j_k}$ for $k=0,1,...,T$
- $\lambda_{k}$ for $k=0,1,...,T$
- $\tau_{k}$ for $k=0,1,...,T$

## Objective 
$$
J=\int_{0}^{T}f(\lambda,\tau)dt=\int_{0}^{T}(\tau^2+\alpha\lambda^2)dt
$$

## Initial condition and terminal condition
- $s_{b_0}=s_b^0\quad s_{b_T}=s_b^T$
- $s_{b_0}=s_b^0\quad s_{b_T}=s_b^T$

## Constraints
- $s_{b_{k+1}}-s_{b_{k}}=\frac{h}{2}(f_{k+1}+f_k)$
- $\tau_k = -J(s_{j_k})^T_{c,j}\lambda_k$
- $s_b^{\min}\leq s_{b_k}\leq s_b^{\max}$
- $s_j^{\min}\leq s_{j_k}\leq s_j^{\max}$
- $\lambda^{\min}\leq \lambda_k\leq \lambda^{\max}$
- $\tau^{\min}\leq \tau_k\leq \tau^{\max}$
